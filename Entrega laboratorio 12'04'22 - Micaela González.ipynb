##### Ejercicio 3.1

Implemente una función que reciba como parámetro una muestra de números (enteros o flotantes) y retorne una lista con los valores *outliers* o *atípicos*. En caso de que no existan valores atípicos, la función debe retornar None.

Los outliers se calculan de la siguiente manera.

$$ outlier < Q_1 - 1.5RIC \vee outlier > Q_3 + 1.5RIC $$

Donde $RIC$ representa el *Rango Intercuartil* y esta dado por $RIC = Q_3 - Q_1$.

El índice del elemento del cuartil $k$ puede calcularse como,

$$ Índice(Q_k) = \frac{k(longitudDatos+1)}{4}  $$

````python
### Resolución ejercicio 3.1
import numpy 
def outliers(parametros):
    outliers=[]
    
    #Ordeno el conjunto
    parametros=sorted(parametros)
    
    #Calculo primer y tercer cuartil
    Q1=numpy.percentile(parametros,25)
    Q3=numpy.percentile(parametros,75)
    RIC=Q3-Q1
    
    #Calculo los limites
    limiteinf=Q1-(1.5*RIC)
    limitesup=Q3+(1.5*RIC)
    
    #Los valores que estan por encima del limitesup y por debajo del limiteinf los agrego a la lista outliers
    for p in parametros:
        if p<limiteinf or p>limitesup:
            outliers.append(p)
    return outliers 

#PRUEBA
outliers([1,2,3,4,89,5,6,7,98])
    
````

El número $\pi$ puede aproximarse con la siguiente fórmula de aproximación,

\begin{equation*}
\pi = 3+ 4 * \sum _ { n = 2,4,6...; k=0,1,2.. } ^ { N,K } (-1)^k \left[ \frac {1}{n (n+1) (n+2)} \right]
\end{equation*}

Implemente una función que reciba como argumento la cantidad de aproximaciones *$N$* a realizar y retorne el valor de la aproximación.

La implementación de la función debe contener anotaciones y documentación apropiada.

NOTA: El valor de $k$ sirve para alternar el signo del término $(-1)^k$.

````python
## Resolución Ejercicio 3.5
def aproxpi(n):
    segundotermino = list(map(lambda n,k : (4*((-1)**k)(1/(n*(n+1)*(n+2)))) , list(range (2, (2*n+1), 2),list(range (0, n)))) #Opero con dos listas, una con valores que comienzan en 2, van de dos en dos y llegan hasta 2*n+1; la otra va desde 0 hasta n. La primer lista corresponde a los valores n y la segunda  alos valores k
    primertermino = 3  # Tomo como primer termino pi= 3 de la formula                    
    for j in segundotermino: #Tomo como segundo termino todo lo que queda por delante del 3 en la formula
        primertermino += j #A pi=3 le voy sumando cada una de las iteraciones n veces realizadas para obtener la aproximacion de pi
    return primertermino
                          
#NOTA: Error de indentación que no supe sortear, pero considero que la lógica está bien
````
